# Выделение цвета

требуется выделить в изображении цвет

Определить диапазон цветов, а затем применить маску к изображению

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image_bgr = cv2.imread(r"C:\Users\user\Desktop\plane.jpg")
```

```python
#конвертировать BGR в HSV
image_hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)
```

```python
#определить диапазон синих значений в HSV
lower_blue = np.array([50, 100, 50])
upper_blue = np.array([130, 255, 255])
```

```python
#создать маску
mask = cv2.inRange(image_hsv, lower_blue,  upper_blue)
```

```python
#наложить маску на изображение
image_bgr_masked = cv2.bitwise_and(image_bgr, image_bgr, mask=mask)
```

```python
#конвертировать BGR в RGB
image_rgb = cv2.cvtColor(image_bgr_masked, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
plt.axis("off")
plt.show
```

```python
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis("off")
plt.show
```

# Удаление фонов

```python
#требуется выделить на изображении передний план
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image_bgr = cv2.imread(r"C:\Users\user\Desktop\plane.jpg")
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
```

```python
#Значения прямоугольника: начальная X, начальная Y, ширина, высота
rectangle = (0, 100, 1000, 300)
```

```python
#создать первоначальную маску
mask = np.zeros(image_rgb.shape[:2], np.uint8)
```

```python
#создать временные массивы, используемые в grabCut
bgModel = np.zeros((1, 65), np.float64)
fgModel = np.zeros((1, 65), np.float64)
```

выполнить алгоритм GrabCut

```python
cv2.grabCut(image_rgb, # наше изображение
            mask,      #маска
            rectangle, #наш прямоугольник
            bgModel,   #временный массив для фона
            fgModel,   #временный массив для переднего плана
            5,         #количество итераций
            cv2.GC_INIT_WITH_RECT) #инициализировать, исполбзуя прямоугольник
```

```python
#создать маску, где фоны уверенно или потенциально равны 0, иначе 1
mask_2 = np.where((mask==2) | (mask==0), 0, 1).astype('uint8')
```

```python
#умножить изображение на новую маску, чтобы вычесть фон
image_rgb_nobg = image_rgb * mask_2[:, :, np.newaxis]
```

```python
plt.imshow(image_rgb_nobg)
plt.axis("off")
plt.show
```

```python
#Показать маску
plt.imshow(mask, cmap='gray')
plt.axis("off")
plt.show
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#установить пороговое значение на одно стандартное отклонение
#выше и ниже медианы интенсивности
lower_threshold = int(max(0, (1.0 - 0.33) *median_intensity))
upper_threshold = int(min(255, (1.0 + 0.33) *median_intensity))
```

# Обнаружение углов

https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_feature2d/py_features_harris/py_features_harris.html

https://habr.com/ru/post/244541/

Требуется обнаружить углы изображения

Использовать реализацию детектора углов Харриса cornerHarris из библиотеки OpenCV

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#загрузить изображение в оттенках серого
image_bgr = cv2.imread(r"C:\Users\user\Desktop\plane.jpg")
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
image_gray = np.float32(image_gray)
```

```python
#задать параметры детектора углов
block_size = 2
aperture = 29
free_parameter = 0.04
```

```python
#обнаружить углы
detector_responses = cv2.cornerHarris(image_gray,
                                      block_size,
                                      aperture,
                                      free_parameter)
```

```python
#крупные угловые маркеры
detector_responses = cv2.dilate(detector_responses, None)
```

```python
#оставить только те отклики детектора, которые больше порога, пометить белым цветом
threshold = 0.02
image_bgr[detector_responses > threshold * detector_responses.max()] = [255, 255, 255]
```

```python
detector_responses
```

```python
threshold * detector_responses.max()
```

```python
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_gray, cmap='gray')
plt.axis("off")
plt.show
```

```python
plt.imshow(detector_responses, cmap='gray')
plt.axis("off")
plt.show()
```

детектор углов Ши-Томази - goodFeaturesToTrack

```python
image_bgr = cv2.imread(r"C:\Users\user\Desktop\plane.jpg")
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
```

```python
#количество углов для обнаружения
corners_to_detect = 10
minimum_quality_score = 0.05
minimum_distance = 25
```

```python
#обнаружить углы
corners = cv2.goodFeaturesToTrack(image_gray,
                                  corners_to_detect,
                                  minimum_quality_score,
                                  minimum_distance)
corners = np.float32(corners)
```

```python
#Нарисовать белый круг в каждом углу
for corner in corners:
    x, y = corner[0]
    cv2.circle(image_bgr, (x,y), 10, (255, 255, 255), -1)
```

```python
#конвертировать в оттенки серого
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_rgb, cmap='gray')
plt.axis("off")
plt.show()
```

# Кодирование гистограмм цветовых каналов в качестве признака

Требуется создать набор признаков, представляющих цвета на изображении.

Вычислить гистограммы для каждого цветового канала

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image_bgr = cv2.imread(r"C:\Users\user\Desktop\plane.jpg", cv2.IMREAD_COLOR)
```

```python
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
```

Создать список для значений признаков

```python
features = []
```

Вычислить гистограмму для каждого цветового канала

```python
colors = ("r", "g", "b")
```

Для каждого цветового канала: вычислить гитограмму и добавить в список значений признаков

```python
for i, channel in enumerate(colors):
    histogram = cv2.calcHist([image_rgb], #изображение
                            [i],          #индекс канала
                            None,         #маска отсутствует
                            [256],        #размер гистограммы
                            [0, 256])     #диапазон
```

```python
features.extend(histogram)
```

```python
observation = np.array(features).flatten()
```
